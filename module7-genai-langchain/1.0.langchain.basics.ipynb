### 1. Development using LLM Native API

````python
from openai import AzureOpenAI
from dotenv import load_dotenv  

load_dotenv()  # if using .env for configuration

client = AzureOpenAI(api_version="2023-03-15-preview")  # expects OPENAI_API_KEY in your environment

topic = "LangChain"
response = client.chat.completions.create(
    model="lums-gpt-4.1-mini",
    temperature=0,
    messages=[
        {"role": "user", "content": f"Explain {topic} in 3 short bullet points."}
    ],
)

result = response.choices[0].message.content
print(result)
````

### 2. Developing using Langchain API

````python
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# OPENAI_API_KEY and OPENAI_API_VERSION should be set in your environment or .env file
llm = AzureChatOpenAI(model="lums-gpt-4.1-mini")

# Basic Example 1: call the chat model directly
msg = llm.invoke("What is Python in one sentence?")
print("Example 1:\n", msg.content, "\n")
````

### 3. Creating Basic Chain using Langchain Expression Language (LCEL)

````python
# Basic Example 2: simple prompt + parser chain
prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in 3 short bullet points."
)

chain = prompt | llm | StrOutputParser()

result = chain.invoke({"topic": "LangChain"})
print(result)
````

````python
# Basic Example 3: reuse your existing chain with different topics
for t in ["LangChain", "Prompt Engineering", "Vector Databases"]:
    print(f"Example 3 ({t}):")
    print(chain.invoke({"topic": t}))
    print("-" * 60)
````
